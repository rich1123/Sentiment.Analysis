```python
# Load yfinance table from MySQL
df_yfinance = pd.read_sql('''select * from yfinance where date between "2020-04-14" and "2020-05-15"
                          ORDER BY date asc''', con=engine)
df_yfinance.set_index('date', inplace = True)
df_yfinance
```

```python
# Load news table from MySQL
df_news = pd.read_sql('select * from news where date between "2020-04-14" and "2020-05-15" ', con=engine)
df_news.set_index('date', inplace = True)
df_news
```

```python
# Load sentiment table from MySQL
df_sentiment = pd.read_sql('select * from sent_a where date between "2020-04-14" and "2020-05-15" ', con=engine)
df_sentiment.set_index('date', inplace = True)
df_sentiment
```

```python
# Load twitter table from MySQL
df_twitter = pd.read_sql('select * from twitter where date between "2020-04-14" and "2020-05-15" ', con=engine)
df_twitter.set_index('date', inplace = True)
df_twitter
```

```python
# Load headline table from MySQL
pd.set_option('display.max_colwidth', -1)
df_headlines = pd.read_sql('select * from headlines', con=engine)
df_headlines.set_index('date', inplace = True)
df_headlines
```

```python
# Merge dfs
df_merged = df_yfinance.merge(df_news, left_on='date', right_on='date')
df_merged = df_merged.merge(df_sentiment, left_on='date', right_on='date')
df_merged = df_merged.merge(df_twitter, left_on='date', right_on='date')
df_merged = df_merged.rename({'count': 'tweet_count'}, axis=1)
df_merged
```

### Regression

```python
# Keep only relevant colunns
df_merged = df_merged.filter(['percent_change','total_articles','daily_sentiment_score', 'tweet_count'], axis=1)
df_merged
```

```python
# Compute correlation matrix
corr_matrix = df_merged.corr()
corr_matrix
```

```python
#pip install statsmodels
import statsmodels.formula.api as sm
```

```python
result = sm.ols(formula="percent_change ~ total_articles + daily_sentiment_score + tweet_count", data=df_merged).fit()
```

## Visuals for Headlines per Day
Using pie and scatter plots for data from News API

```python
from math import pi

import pandas as pd

from bokeh.models import Legend, ColumnDataSource
from bokeh.io import output_file, output_notebook, show
from bokeh.palettes import Cividis256
from bokeh.plotting import figure
from bokeh.transform import cumsum
from bokeh.models.tools import HoverTool
from bokeh.transform import factor_cmap



colors = ['#20B2AA', '#2F4F4F', '#008080', '#008B8B', '#00FFFF', '#0000FF', '#B0E0E6', '#191970',
'#20B2AB', '#2F4F4C', '#00808D', '#008B8E', '#00FFFG', '#000GFF', '#B0E7E6', '#191971',
'#20B2A9', '#2F4F48', '#008087', '#008B86', '#00FFF5', '#0000F4', '#B0E0E3', '#191972',
'#20B2AB', '#2F4F4G', '#008082']


# source = hell.rename(columns={'index':'date'})
source = hl.reset_index(name='count').rename(columns={'index':'date'})
source['date']= pd.to_datetime(source['date'])
source['angle'] = source['count']/source['count'].sum() * 2*pi
source['color'] = colors

p = figure(plot_width= 500, plot_height=500, title="Headlines Per Day", toolbar_location=None,
            x_range=(-0.5, 1.0))

px = figure(x_axis_type='datetime', plot_width= 500, plot_height=500, title="Headlines Per Day", toolbar_location=None)

hover = HoverTool(
 tooltips=[("Date", "@date{%m-%d-%Y}"),
           ("Count per Day" , "@count"),
                        ],
    formatters={
        'date':'datetime',
        'count' : 'printf',
},
)

hoverx = HoverTool(
 tooltips=[("Date", "@date{%m-%d-%Y}"),
           ("Count per Day" , "@count"),
                        ],
    formatters={
        'date':'datetime',
        'count' : 'printf',
},
)

p.add_tools(hover)
px.add_tools(hoverx)

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', source=source)
px.circle(x='date', y='count', line_width=3, size=10, source=source,
         color="pink", nonselection_fill_color='green', selection_color='pink')

# px.line(x='index', y='count', line_width=3, source=source)


p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

px.yaxis.axis_label = 'Headlines'
px.xaxis.axis_label = 'Date'

px.y_range.start = 0
px.y_range.end = 20

#output format 
output_notebook()
output_file('headlines__count_plot.html')

# px.legend.location = "center_right"
# # p.legend.formatters={'date': 'datetime'}
# p.legend.click_policy="mute"

layoutp= row(px, p)
show(layoutp)


# top_left', 'top_center', 'top_right', 'center_left',
# 'center', 'center_right', 'bottom_left', 'bottom_center', 'bottom_right
```

## High, Low, and Opening Stock Price for TSLA

```python
df_yfin = df_yfinance.reset_index()
df_yfin['date']= pd.to_datetime(df_yfin['date'])
```

```python

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import column, grid 

source = ColumnDataSource(data=dict(
        date=df_yfin['date'],
        h=df_yfin['high'],
        l=df_yfin['low'],
        o=df_yfin['open'],
    ))

hover = HoverTool (
    tooltips = [
        ("date","@date{%m-%d-%Y}"),
        ("high price", "$@h"),
        ("low price", "$@l"),
    ("opening price", "$@o"),
           ],

formatters={
        'date':'datetime',
},
)

p4 = figure(x_axis_type="datetime", background="blue", plot_height=400, plot_width=500, title="Tesla Stock Prices")
p4.grid.grid_line_alpha=0.3
p4.xaxis.axis_label = 'Date'
p4.yaxis.axis_label = 'Price'

p4.line(x='date', y='h', color='#FB9A99', legend_label='High Price', source=source)
p4.line(x='date', y='l', color='#A6CEE3', legend_label='Low Price',  source=source)
p4.line(x='date', y='o', color='#B2DF8A', legend_label='Opening Price',  source=source)

p4.add_tools(hover)

output_notebook()

show(p4)

# %m/%d/%y
# date time format
```
